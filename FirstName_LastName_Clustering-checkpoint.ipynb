```python
# Import necessary libraries
from sklearn.metrics import davies_bouldin_score, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Evaluation: Clustering Logic and Metrics
def evaluate_clustering_metrics(data, labels, cluster_centers):
    """
    Evaluate clustering using the following metrics:
    1. Davies-Bouldin Index (DB Index)
    2. Silhouette Score (optional, but useful)
    3. Cluster summaries (mean of features in each cluster)
    """
    # 1. Davies-Bouldin Index (DB Index)
    db_index = davies_bouldin_score(data, labels)
    print(f"Davies-Bouldin Index: {db_index:.4f}")
    
    # 2. Silhouette Score (optional but valuable)
    silhouette_avg = silhouette_score(data, labels)
    print(f"Silhouette Score: {silhouette_avg:.4f}")
    
    # 3. Cluster summaries (mean of features in each cluster)
    # Adjusting for column names based on the number of features in your data
    cluster_summary = pd.DataFrame(data, columns=["TotalSpend", "TotalQuantity", "UniqueProducts", "Feature4"])
    cluster_summary["Cluster"] = labels
    print("\nCluster Summaries (Feature Means per Cluster):")
    print(cluster_summary.groupby("Cluster").mean())
    
    return db_index, silhouette_avg, cluster_summary

# Visual Representation: Cluster Plot
def plot_cluster_visualization(data, labels, cluster_centers, feature_names):
    """
    Visualize the clusters in 2D (using PCA for dimensionality reduction) and plot cluster centers.
    """
    # Reduce to 2D using PCA for visualization
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(data)

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], hue=labels, palette="tab10", s=100, alpha=0.6, edgecolor="w")
    
    # Plot cluster centers
    centers_2d = pca.transform(cluster_centers)
    plt.scatter(centers_2d[:, 0], centers_2d[:, 1], c="red", marker="X", s=200, label="Cluster Centers")
    
    plt.title("Customer Segmentation Clusters")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend()
    plt.show()

# Example Data - assuming customer_features_scaled and labels are already defined
# For demonstration, here’s an example of fitting a KMeans model
# customer_features_scaled = ... (your scaled features here)
# Assuming labels are generated from the KMeans model

# KMeans clustering for example
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(customer_features_scaled)

# Run clustering evaluation
db_index, silhouette_avg, cluster_summary = evaluate_clustering_metrics(customer_features_scaled, labels, kmeans.cluster_centers_)

# Visualize clusters and cluster centers
plot_cluster_visualization(customer_features_scaled, labels, kmeans.cluster_centers_, ["TotalSpend", "TotalQuantity", "UniqueProducts", "Feature4"])
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customers = pd.read_csv("Customers.csv")
transactions = pd.read_csv("Transactions.csv")

# Merge datasets to create a unified view of customer profiles and transactions
merged_data = transactions.merge(customers, on="CustomerID")

# Feature Engineering: Aggregate transaction data per customer
customer_features = merged_data.groupby("CustomerID").agg(
    TotalSpend=("TotalValue", "sum"),
    TotalQuantity=("Quantity", "sum"),
    UniqueProducts=("ProductID", lambda x: x.nunique())
).reset_index()

# Add profile information
customer_features = customer_features.merge(customers, on="CustomerID")

# Encode categorical features (Region)
customer_features["Region"] = customer_features["Region"].astype("category").cat.codes

# Normalize numerical features
scaler = StandardScaler()
numerical_features = ["TotalSpend", "TotalQuantity", "UniqueProducts"]
customer_features_scaled = scaler.fit_transform(customer_features[numerical_features + ["Region"]])

# Clustering using KMeans
def perform_clustering(data, n_clusters):
    # Initialize and fit KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data)
    
    # Calculate Davies-Bouldin Index
    db_index = davies_bouldin_score(data, labels)
    
    return labels, db_index, kmeans

# Determine optimal number of clusters (2 to 10)
db_indices = []
for n_clusters in range(2, 11):
    _, db_index, _ = perform_clustering(customer_features_scaled, n_clusters)
    db_indices.append(db_index)

# Plot DB Index for each number of clusters
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), db_indices, marker="o")
plt.title("DB Index vs. Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Davies-Bouldin Index")
plt.xticks(range(2, 11))
plt.grid()
plt.show()

# Perform clustering with the optimal number of clusters (lowest DB Index)
optimal_clusters = np.argmin(db_indices) + 2
labels, db_index, kmeans = perform_clustering(customer_features_scaled, optimal_clusters)

print(f"Optimal Number of Clusters: {optimal_clusters}")
print(f"Davies-Bouldin Index for Optimal Clusters: {db_index:.4f}")

# Add cluster labels to customer data
customer_features["Cluster"] = labels

# Visualize clusters using a pairplot (first two numerical features)
sns.pairplot(
    customer_features,
    vars=["TotalSpend", "TotalQuantity", "UniqueProducts"],
    hue="Cluster",
    palette="tab10",
    diag_kind="kde",
    corner=True
)
plt.suptitle("Customer Segments", y=1.02)
plt.show()

# Save clustering results to a CSV file
customer_features.to_csv("Nikita_Chauhan_Clustering.csv", index=False)

# Display summary of clusters
print(customer_features.groupby("Cluster").mean())
```
